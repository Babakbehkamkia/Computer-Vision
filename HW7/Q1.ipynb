# Panorama with OpenCV

```python
import numpy as np
import cv2
import glob
import matplotlib.pylab as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')

path = "/content/drive/My Drive/CV_course/HW7/images/"
```

```python
# read input victoria images and show them in a row together
from matplotlib.pyplot import figure, imshow, axis
from matplotlib.image import imread

image_list = []
fig = figure(figsize=(30,15))
for i in range(1,8):
    a=fig.add_subplot(1,7,i)
    image = cv2.imread(path + f"victoria{i}.png")
    image_list.append(image)
    imshow(image,cmap='Greys_r')
    axis('off')
```

```python
# initialize OpenCV's image sticher object and then perform the image stitching on input images
stitchy=cv2.Stitcher.create()
(dummy,output)=stitchy.stitch(image_list)
```

```python
# show victoria panorama
from google.colab.patches import cv2_imshow

cv2_imshow(output)
```
